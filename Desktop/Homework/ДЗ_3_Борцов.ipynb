Создайте SQLite базу данных и таблицу через sqlite3 , заполните её несколькими строками и извлеките их с помощью pd.read_sql().

````python
import sqlite3
````

````python
conn = sqlite3.connect('example.db')
cursor = conn.cursor()

cursor.execute('''
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER,
    grade TEXT
)
''')

cursor.execute("DELETE FROM students")
cursor.execute("INSERT INTO students (name, age, grade) VALUES ('Иван', 20, 'A')")
cursor.execute("INSERT INTO students (name, age, grade) VALUES ('Мария', 22, 'B')")
cursor.execute("INSERT INTO students (name, age, grade) VALUES ('Алексей', 21, 'A')")

conn.commit()
df_sql = pd.read_sql('SELECT * FROM students', conn)
print(df_sql)
conn.close()
````

````python
DF['date_str'] = ['2022-02-13',
                  '2025-01-12',
                  '2024-11-09',
                  '2024-02-28',
                  '2023-09-11',
                  '2024-08-14',
                  '2024-12-13',
                  ]
DF
````

````python
dates = []
for date_str in DF["date_str"]:
    date = datetime.strptime(date_str, '%Y-%m-%d')
    dates.append(date)

ym = []
for date in dates:
    x = {
        "Year": date.year,
        "Month": date.month
    }
    ym.append(x)
    print(x)
````
